```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "constituents-financials_csv.csv"
financial_df = pd.read_csv(csv_file)
financial_df.head()
```

```python
new_financial_df = financial_df[['Symbol', 'Name', 'Sector', 'Price', 'Price/Earnings', 'Dividend Yield',\
                                '52 Week Low', '52 Week High']].copy()
new_financial_df.head()
```

```python
new_financial_df["52 Week Diff"] = new_financial_df["52 Week Low"] - new_financial_df["52 Week High"]
new_financial_df.head()
```

```python
rds_connection_string = "postgres:changeme@localhost:5432/postgres"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
conn = engine.connect()
```

```python
conn.execute("commit")
```

```python
conn.execute("CREATE DATABASE financial_db")
```

```python
conn.close()
```

```python
rds_connection_string = "postgres:changeme@localhost:5432/financial_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
new_financial_df.to_sql(name="financial", con=engine, if_exists='append', index=False)
```
